```python
import pandas as pd
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# bad example first
# Load the CSV file for 3 individuals
csv_path = 'labeling_results_bad.csv'  # Update the path if necessary
df = pd.read_csv(csv_path)
labels = ['T1', 'T2', 'P2']
# True labels (A is always 'A', B is always 'B', C is always 'C')
true_labels = ['A'] * len(df) + ['B'] * len(df) + ['C'] * len(df) 

# Predicted labels are extracted from the corresponding columns
predicted_labels = df['true_label_A'].tolist() + df['true_label_B'].tolist() + df['true_label_C'].tolist()

# Generate the confusion matrix for all combined labels
cm = confusion_matrix(true_labels, predicted_labels, labels=['A','B','C'])

plt.figure(figsize=(10, 8))  # Adjust the figure size
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=labels, yticklabels=labels,
            annot_kws={"size": 16})  # Increased annotation font size

plt.xlabel("Predicted", fontsize=16)  # Increase x-axis label font size
plt.ylabel("True", fontsize=16)  # Increase y-axis label font size
plt.xticks(fontsize=14)  # Increase x-tick label font size
plt.yticks(fontsize=14)  # Increase y-tick label font size
plt.savefig('plots/confusion_matrix_bad_example.png', dpi=150)
plt.show()


# good example second
# Load the CSV file for 3 individuals
csv_path = 'labeling_results_good.csv'  # Update the path if necessary
df = pd.read_csv(csv_path)
labels = ['T1', 'P1', 'P2', 'P3']
# True labels (A is always 'A', B is always 'B', C is always 'C')
true_labels = ['A'] * len(df) + ['B'] * len(df) + ['C'] * len(df) + ['D'] * len(df)

# Predicted labels are extracted from the corresponding columns
predicted_labels = df['true_label_A'].tolist() + df['true_label_B'].tolist() + df['true_label_C'].tolist()+ df['true_label_D'].tolist()

# Generate the confusion matrix for all combined labels
cm = confusion_matrix(true_labels, predicted_labels, labels=['A','B','C','D'])

plt.figure(figsize=(10, 8))  # Adjust the figure size
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=labels, yticklabels=labels,
            annot_kws={"size": 16})  # Increased annotation font size

plt.xlabel("Predicted", fontsize=16)  # Increase x-axis label font size
plt.ylabel("True", fontsize=16)  # Increase y-axis label font size
plt.xticks(fontsize=14)  # Increase x-tick label font size
plt.yticks(fontsize=14)  # Increase y-tick label font size
plt.savefig('plots/confusion_matrix_good_example.png', dpi=150)
plt.show()
```
